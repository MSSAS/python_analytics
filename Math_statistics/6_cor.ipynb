```python
import pandas as pd
import seaborn as sns 
import numpy as np
import pingouin as pg
import scipy.stats as ss
from sklearn.datasets import fetch_openml
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.select_dtypes(include='number').corr()
```

```python
sns.scatterplot(data=tips, x="total_bill", y='tip')
```

```python
#Корреляция Пирсона
ss.pearsonr(tips.total_bill, tips.tip)
```

```python
#Корреляция Спирмена.
#Используется, когда данные не следуют нормальному распределению или когда предполагаемая зависимость может быть нелинейной

ss.spearmanr(tips.total_bill, tips.tip)
```

```python
sns.regplot(data=tips, x="total_bill", y='tip')
```
